```python
import numpy as np
import pandas as pd

raw_data_movie = pd.read_csv("movies.csv")
raw_data_rating = pd.read_csv("ratings.csv")
```

```python
display(raw_data_movie.head())
raw_data_rating.head()
```

```python
rated_movieId_list = raw_data_rating["movieId"].unique()

raw_data_movie = raw_data_movie[raw_data_movie["movieId"].isin(rated_movieId_list)]
raw_data_movie
```

```python
genres_tag_dict = {}
genres_list = []
tag_list = []

for s in raw_data_movie.genres.str.split('|'):
    if type(s) == type(0.1):
        continue
    for g in s:
        genres_tag_dict[g] = 0

for key, val in genres_tag_dict.items():
    genres_list.append(key)

tag_list += genres_list
genres_list
```

```python
for g in genres_list:
    raw_data_movie[g] = np.where(raw_data_movie.genres.str.find(g) != -1, "O", "-")

data_movie = raw_data_movie.drop(columns=["genres"])
data_movie = data_movie.set_index("movieId")
data_movie
```

```python
data_merge = pd.merge(raw_data_rating, data_movie["title"], on="movieId")
data_merge
```

```python
rating_tbl = data_merge.pivot_table(values="rating", index="title", columns="userId")

for col, item in rating_tbl.items() :
    rating_tbl[col] = item.fillna(0)
    
rating_tbl
```

```python
user_count, movie_count = rating_tbl.shape

from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=12)
matrix = SVD.fit_transform(rating_tbl)

matrix.shape
```

```python
corr = np.corrcoef(matrix)
corr.shape
```

```python
movie_title_list = list(rating_tbl.index)
movie_title_list
```

```python
search_recommanded = input("Input Movie Title : ")

idx = movie_title_list.index(search_recommanded)
similarity = corr[idx]

idx = 0
idx_dict = {}
for val in similarity:
    if val >= 0.9:
        idx_dict[val] = movie_title_list[idx]
    idx += 1

movie_titles = []
limit = 10
idx = 0
for key, val in sorted(idx_dict.items(), reverse=True):
    if limit < idx:
        break

    if idx == 0:
        idx += 1
        continue

    movie_titles.append(val)
    idx += 1

search_table = data_movie[data_movie["title"].isin(movie_titles)]

search_table
```
